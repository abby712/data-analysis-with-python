#Dataset Information

Source: Kaggle

Year / Timeline: 2025

Dataset Description:

MEMBER_NO: Unique identification number assigned to each loyalty program member.
FFP_DATE: Date when the customer joined the Frequent Flyer Program

FIRST_FLIGHT_DATE: Date of the customer’s first recorded flight with the airline.

GENDER: Gender of the customer.
FFP_TIER	Membership level or loyalty tier based on flight activity or spending.

WORK_CITY: City where the customer works or resides.


WORK_PROVINCE: Province/state of the customer’s workplace or residence.

WORK_COUNTRY: Country of the customer’s workplace or residence.

AGE: Age of the customer.

LOAD_TIME: Date when the customer’s record was last updated or loaded into the system.

FLIGHT_COUNT: Total number of flights taken by the customer during the observed period.

BP_SUM: Total bonus points  accumulated from flight activity.

SUM_YR_1: Total revenue generated by the customer in the first observed year.

SUM_YR_2: Total revenue generated by the customer in the second observed year.

SEG_KM_SUM: Total flight distance traveled by the customer.

LAST_FLIGHT_DATE: Date of the customer’s most recent flight.

LAST_TO_END: Number of days between the last flight date and the data extraction.

AVG_INTERVAL: Average number of days between consecutive flights for each customer.

MAX_INTERVAL: Maximum number of days between two consecutive flights.

EXCHANGE_COUNT: Number of times the customer exchanged or redeemed loyalty points.

avg_discount: Average flight discount rate received by the customer.

Points_Sum: Total loyalty points earned by the customer.

Point_NotFlight: Number of loyalty points earned from non-flight activities.

## Stage 2 – Data Cleaning and Pre-processing

1. Handle missing values (impute or drop)
2. Changing Negative values to postive values
3. Handle duplicates
4. Treat outliers if required
5. Standardizing columns
6. Convert data types if needed
7. Rounding the Decimal numer to its nearest integer
8. Checking for skewness
9. Skewness Transformation( Derived fields required for analysis)
10. Skewness Comaprison Visuals




```python

# Before Cleaning

print("Duplicates before:", df.duplicated().sum())

print("\nMissing values before:")
print(df.isnull().sum())

total_missing_before = df.isnull().sum().sum()
print("Total Missing values:",total_missing_before )


print("\nData Types before:")
print(df.dtypes)
```

```python
# Converting Object data type column to its original data type

df['SUM_YR_1'] = pd.to_numeric(df['SUM_YR_1'], errors='coerce')
df['SUM_YR_1'] = df['SUM_YR_1'].astype('float')
```

```python
# Handling Missing values- Text columns

# i) Find Object type columns
text_cols = df.select_dtypes(include=['object']).columns
print("Text columns:", text_cols)
# ii) Fill missing values
df[text_cols] = df[text_cols].fillna('Unknown')

df[text_cols] = df[text_cols].replace('Unknown', df[text_cols].mode().iloc[0])
```

```python
# Handling Missing values- Numeric columns

# i) Find Numeric type columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("Numeric columns:", numeric_cols)

# ii) Fill Missing values

for col in numeric_cols:
  df[col].fillna(df[col].max(),inplace=True)
```

```python
df.isnull().sum().sum()
```

```python
# Changing Negative values to postive values

df[df.select_dtypes(include='number').columns] = (
    df.select_dtypes(include='number').abs())
```

```python
# Outlier Handling

# Select only numeric columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Checking for outliers
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identifying outliers
outliers = ((df[num_cols] < lower_bound) | (df[num_cols] > upper_bound))
outlier_count = outliers.sum()

print("Outliers Detected in Each Column (Before Cleaning):\n")
print(outlier_count)

# Before Cleaning visual

plt.figure(figsize=(12, 6))
sns.boxplot(data=df[num_cols], orient='h', palette='Set2')
plt.title("Boxplot Showing Outliers Before Cleaning")
plt.ticklabel_format(style='plain', axis='x')
plt.xlabel("Value Range")
plt.ylabel("Columns")
plt.tight_layout()
plt.show()


# Handle outliers by capping with lower/upper bounds (IQR method)
for col in num_cols:
    df.loc[df[col] < lower_bound[col], col] = lower_bound[col]
    df.loc[df[col] > upper_bound[col], col] = upper_bound[col]

# Outliers checking after cleaning
Q1_clean = df[num_cols].quantile(0.25)
Q3_clean = df[num_cols].quantile(0.75)
IQR_clean = Q3_clean - Q3_clean

lower_bound_clean = Q1_clean - 1.5 * IQR_clean
upper_bound_clean = Q3_clean + 1.5 * IQR_clean

outliers_after = ((df[num_cols] < lower_bound_clean) | (df[num_cols] > upper_bound_clean))
outlier_count_after = outliers_after.sum()

print("\n Outliers After Cleaning:\n")
print(outlier_count_after)

#After Cleaning visual

plt.figure(figsize=(12, 6))
sns.boxplot(data=df[num_cols], orient='h', palette='Set1')
plt.title("Boxplot Showing Outliers After Cleaning")
plt.ticklabel_format(style='plain', axis='x')
plt.xlabel("Value Range")
plt.ylabel("Columns")
plt.tight_layout()
plt.show()


```

```python
# Standardizing columns

obj_columns = ["GENDER", "WORK_CITY", "WORK_PROVINCE", "WORK_COUNTRY"]
for col in obj_columns:
    df[col] = df[col].astype(str).str.strip()

df["WORK_CITY"] = df["WORK_CITY"].str.title()
df["WORK_PROVINCE"] = df["WORK_PROVINCE"].str.title()
df["WORK_COUNTRY"] = df["WORK_COUNTRY"].str.title()
df["GENDER"] = df["GENDER"].str.title()


# Gender Column
df["GENDER"] = df["GENDER"].replace(["Male","Mal","Mle","M","male"], "m")
df["GENDER"] = df["GENDER"].replace("m", "M")
df["GENDER"] = df["GENDER"].replace(["female","Fem","Feme","F","Female"], "f")
df["GENDER"] = df["GENDER"].replace("f", "F")
df["GENDER"] = df["GENDER"].replace("Nan", "Unknown")
# Filling missing values with the most frequent value
df['GENDER'] = df['GENDER'].replace("Unknown", df['GENDER'].mode()[0])
print(df["GENDER"].value_counts())

# Work_Country Column
df["WORK_COUNTRY"] = df["WORK_COUNTRY"].replace(["India","In"], "IN")
df["WORK_COUNTRY"] = df["WORK_COUNTRY"].replace("IN", "India")
df["WORK_COUNTRY"] = df["WORK_COUNTRY"].replace(["Cn","CHN","Chn","Cn-Prc","China Mainland"], "China")
df["WORK_COUNTRY"] = df["WORK_COUNTRY"].replace("Us", "United States")
df["WORK_COUNTRY"] = df["WORK_COUNTRY"].replace("Unknown", df["WORK_COUNTRY"].mode()[0])

# Work_Province Column
df["WORK_PROVINCE"] = df["WORK_PROVINCE"].replace("Bei Jing", "Beijing")
df["WORK_PROVINCE"] = df["WORK_PROVINCE"].replace("Guang Dong", "Guangdong")
df["WORK_PROVINCE"] = df["WORK_PROVINCE"].replace("Zhejiang", "Zhe Jiang")
df["WORK_PROVINCE"] = df["WORK_PROVINCE"].replace("Shang Hai", "Shanghai")
df["WORK_PROVINCE"] = df["WORK_PROVINCE"].replace("Sichuan", "Si Chuan")
df["WORK_PROVINCE"] = df["WORK_PROVINCE"].replace("Shaanxi", "Shanghai")
df["WORK_PROVINCE"] = df["WORK_PROVINCE"].replace(["Unknown"], df["WORK_PROVINCE"].mode()[0])

# Work_City Column
df["WORK_CITY"] = df["WORK_CITY"].replace("Chongqing", "Chong Qing")
df["WORK_CITY"] = df["WORK_CITY"].replace("Xi'An", "Xian")
df["WORK_CITY"] = df["WORK_CITY"].replace("Nan Jing", "Nanjing")
df["WORK_CITY"] = df["WORK_CITY"].replace(["Unknown"], df["WORK_CITY"].mode()[0])
```

```python
#  Removing Duplicates
df.drop_duplicates(inplace=True)
```

```python
# Changing Date columns format(Object to Date)

d_columns = ['FFP_DATE', 'FIRST_FLIGHT_DATE', 'LOAD_TIME', 'LAST_FLIGHT_DATE']
df[d_columns] = df[d_columns].apply(pd.to_datetime, errors='coerce', infer_datetime_format=True)

# Filling in missing values of Date columns
d_columns = ['FFP_DATE', 'FIRST_FLIGHT_DATE', 'LOAD_TIME', 'LAST_FLIGHT_DATE']
for col in d_columns:
      df[col] = df[col].fillna(df[col].median())
```

```python
# Changing Data type from float to integer

f_columns = df.select_dtypes(include=['float64']).columns.tolist()
print(f_columns)

df[f_columns] = df[f_columns].fillna(0).round().astype('int64')
```

```python
# Skewness Checking

# Selecting numeric columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns


# Applying transformations

df['SUM_YR_1'] = pd.to_numeric(df['SUM_YR_1'], errors='coerce')
df['SUM_YR_2'] = pd.to_numeric(df['SUM_YR_2'], errors='coerce')
df['AVG_INTERVAL'] = pd.to_numeric(df['AVG_INTERVAL'], errors='coerce')


# Calculating skewness
skewness = df[numeric_cols].skew().sort_values(ascending=False)
print("\n Calculated Skewness")
print(skewness)


# Applying Transformations safely
df['SUM_YR_1_log'] = np.log1p(df['SUM_YR_1'])
df['SUM_YR_2_log'] = np.log1p(df['SUM_YR_2'])
df['AVG_INTERVAL_sqrt'] = np.sqrt(df['AVG_INTERVAL'])
```

```python
# Skewness comparison visuals
# 1.Original SUM_YR_1
plt.subplot(2, 2, 1)
sns.histplot(df['SUM_YR_1'], kde=True, color='skyblue')
plt.title("Original SUM_YR_1")
plt.xlabel("SUM_YR_1")
plt.ylabel("Frequency")

# 2.Log Transformed SUM_YR_1
plt.subplot(2, 2, 2)
sns.histplot(df['SUM_YR_1_log'], kde=True, color='orange')
plt.title("Log Transformed SUM_YR_1")
plt.xlabel("SUM_YR_1_log")
plt.ylabel("Frequency")

# 3.Original SUM_YR_2
plt.subplot(2, 2, 3)
sns.histplot(df['SUM_YR_2'], kde=True, color='lightgreen')
plt.title("Original SUM_YR_2")
plt.xlabel("SUM_YR_2")
plt.ylabel("Frequency")

# 4.Log Transformed SUM_YR_2
plt.subplot(2, 2, 4)
sns.histplot(df['SUM_YR_2_log'], kde=True, color='salmon')
plt.title("Log Transformed SUM_YR_2")
plt.xlabel("SUM_YR_2_log")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# 1.Count Plot

sns.countplot(data=df, x='GENDER', color='skyblue')
plt.title("Customer Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

## Interpretation of your above chart

The count plot displays the number of customers in each gender category using bars.

It visually compares how many male and female customers exist in the dataset. The taller the bar, the more customers that belong to that gender.

GENDER column is used to group and count customers by gender because
It’s a categorical variable suitable for understanding demographic composition.

 We can see the gender composition of all customers, which helps understand demographic diversity or imbalance in the dataset.





```python
# 2.Grouped Bar chart

grouped = df.groupby(['WORK_COUNTRY', 'FFP_TIER']).size().unstack()
grouped.plot(kind='bar', stacked=True, figsize=(8,5))
plt.title("FFP Tier Distribution by Country")
plt.xlabel("Work Country")
plt.ylabel("Number of Customers")
plt.show()
```

## Interpretation of your above chart

The Grouped bar chart compares the number of customers in different Frequent Flyer Program (FFP) tiers across multiple work countries.

It highlights how customer membership levels like Tier 1, Tier 2, Tier 3 are distributed across countries and also identifies which countries have more high-tier or low-tier members.

WORK_COUNTRY is used for grouping customers by country.
FFP_TIER is used to show membership tier levels within each country.


we can see the cross country distribution of loyalty tiers, helping identify which countries have stronger engagement or premium-tier customers.


```python
# 3.Donut Chart

bins = [0, 18, 30, 45, 60, 100]
labels = ['Teen', 'Young Adult', 'Adult', 'Middle Age', 'Senior']

df['AGE_GROUP'] = pd.cut(df['AGE'], bins=bins, labels=labels)
age_group = df['AGE_GROUP'].value_counts().sort_index()

plt.pie(age_group, labels=labels, autopct='%1.1f%%', wedgeprops={'width':0.5})
plt.title("Customer Age Group Composition")
plt.show()
```

## Interpretation of your above chart

The Donut chart is a variation of a pie chart, showing proportions of customer age groups.

It shows what percentage of customers fall into different age categories such as Teen, Young Adult, Adult, Middle Age, and Senior.

AGE column is used to classify customers into age ranges.

We can see the demographic spread of customer ages, providing insight into which age group dominates the customer base.


```python
# 4.Pie plot
tier_counts = df['FFP_TIER'].value_counts()
plt.pie(tier_counts, labels=tier_counts.index,autopct='%1.1f%%')
plt.title("Customer Distribution by FFP Tier")
plt.show()
```

## Interpretation of your above chart

The pie chart shows how customers are distributed across different FFP tiers.

The chart highlights which membership tier has the most or least customers.

FFP_TIER column is used to group customers by membership level.

We can see the overall membership structure of the customer base by a quick visual summary of loyalty tier participation.

```python
# 5.Scatter Plot

df['Distance_Group'] = pd.cut(df['SEG_KM_SUM'],
                              bins=[0, 5000, 20000, 50000, 100000, 200000, 500000],
                              labels=['0-5k','5k-20k','20k-50k','50k-100k','100k-200k','200k+'])

grouped = df.groupby('Distance_Group')['Points_Sum'].mean().reset_index()
plt.figure(figsize=(7,5))
sns.scatterplot(x='Distance_Group', y='Points_Sum', data=grouped, s=100, color='coral')
plt.plot(grouped['Distance_Group'], grouped['Points_Sum'], color='gray', linestyle='--')
plt.title("Average Points by Distance Range")
plt.xlabel("Distance Range (KM)")
plt.ylabel("Average Points Earned")
plt.show()
```

## Interpretation of your above chart

 This chart combines a scatter plot and the line plot to show the average points earned by customers in different distance ranges.Each point represents the average points for customers who have flown within a specific total distance range.

The chart shows the relationship between total flight distance and loyalty points earned we can observe How average points change as flight distance increases and reveals whether flying more often leads to earning more points.

SEG_KM_SUM column is used to show total distance flown by each customer.

We can see the relationship between how far customers fly and how many points they earn on average and also Understand whether long-distance travelers earn more loyalty points.


```python
# 6.Heatmap

plt.figure(figsize=(8,6))
cols = ['BP_SUM', 'SUM_YR_1', 'SUM_YR_2', 'Points_Sum', 'Point_NotFlight', 'avg_discount']
corr2 = df[cols].corr().round(2)

sns.heatmap(corr2, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title("Correlation: Spending vs Points Behavior")
plt.show()

corr2.to_csv("correlation_matrix.csv")
from google.colab import files
files.download("correlation_matrix.csv")

```

## Interpretation of your above chart

The Correlation heatmap shows the relationships between spending, points, and discount-related features of customers.Each cell in the heatmap represents how strongly two variables are positively or negatively correlated with each other.

The chart shows how different spending and points-related features are connected.we can see Which factors are strongly related and Which features have weak or no correlation.

BP_SUM to show Total bonus points earned.
SUM_YR_1 and SUM_YR_2 to see Total money spent or ticket fare in two different years.
Points_Sum to see Total points earned.
Point_NotFlight to show Points earned from non-flight activities.
avg_discount to see Average discount customers received.

We can see the relationship between customer spending and point accumulation patterns that helps us to Identify which spending factors have the strongest influence on earning points and Understand whether discounts affect total points or spending behavior.We also Gain insights into how financial activities and loyalty points are connected.
